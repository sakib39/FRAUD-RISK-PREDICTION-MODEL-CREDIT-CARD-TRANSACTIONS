### Time Distribution (When Are Transactions Happening?)

```python
plt.figure(figsize=(6,4))
df['Time'].hist(bins=50)
plt.title('Transaction Time Distribution')
plt.xlabel('Time (seconds since first transaction)')
plt.ylabel('Frequency')
plt.show()
```

`Time` represents the number of seconds since the first transaction in the dataset.

The histogram shows how transaction volume changes over the two-day period:

- Peaks indicate **busy periods** with many transactions.
- Valleys indicate **quieter periods**.

Later, we can engineer time-based features (e.g., **hour-of-day**, **time-of-day buckets**) to capture behavioral patterns that may differ between fraud and non-fraud transactions.

### Transactions by Relative 6-Hour Windows

```python
df['Hour'] = (df['Time'] / 3600) % 24

def get_time_window(hour):
    if 0 <= hour < 6:
        return 'Window_1 (0â€“6 hrs)'
    elif 6 <= hour < 12:
        return 'Window_2 (6â€“12 hrs)'
    elif 12 <= hour < 18:
        return 'Window_3 (12â€“18 hrs)'
    else:
        return 'Window_4 (18â€“24 hrs)'

df['TimeWindow'] = df['Hour'].apply(get_time_window)

plt.figure(figsize=(6,4))
sns.countplot(data=df, x='TimeWindow', order=[
    'Window_1 (0â€“6 hrs)',
    'Window_2 (6â€“12 hrs)',
    'Window_3 (12â€“18 hrs)',
    'Window_4 (18â€“24 hrs)'
])
plt.title('Transactions by Relative 6-Hour Windows')
plt.xlabel('Time Window (Relative)')
plt.ylabel('Number of Transactions')
plt.show()
```



The `Time` column measures **seconds since the first recorded transaction**, not real-world clock time.
Therefore, we interpret it as a **relative time** variable.

To analyze temporal patterns, we divide each 24-hour cycle into **four 6-hour windows**:
- 0â€“6 hrs  
- 6â€“12 hrs  
- 12â€“18 hrs  
- 18â€“24 hrs  

#### ðŸ§  Observations
- Transaction frequency fluctuates across these relative windows.
- Fraudulent transactions may cluster in **low-activity periods** within the two-day cycle.

#### ðŸ’¼ Insight
Even without absolute timestamps, relative time still provides useful behavioral patterns.  
These **periodic activity cycles** can help the model capture temporal dynamics related to fraud behavior.

## Building the Fraud Risk Prediction Model

### Data Preprocessing and Train-Test Split

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
df['scaled_amount'] = scaler.fit_transform(df['Amount'].values.reshape(-1,1))
```

```python
df['scaled_time'] = scaler.fit_transform(df['Time'].values.reshape(-1,1))
```

```python
df = df.drop(['Time','Amount'], axis=1)
```

```python
X = df.drop('Class', axis=1)
y = df['Class']
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)
```

We scale continuous variables (`Amount`, `Time`) to normalize their ranges and remove potential bias from large numerical differences.  

After that, we separate features (`X`) and labels (`y`), and perform a **stratified split** to preserve the fraud/non-fraud ratio across training and test sets.


### Handling Class Imbalance with SMOTE

```python
from imblearn.over_sampling import SMOTE
sm = SMOTE(random_state=42)
X_res, y_res = sm.fit_resample(X_train, y_train)
```

```python
print("Before SMOTE:", y_train.value_counts())
print("After SMOTE:", y_res.value_counts())
```

Because fraudulent transactions are extremely rare (~0.17%), we apply **SMOTE (Synthetic Minority Over-sampling Technique)** to balance the classes.  

This generates synthetic fraud samples based on nearest neighbors, improving the modelâ€™s ability to detect fraud patterns without overfitting to exact duplicates.

### Baseline Model Training

#### **Logistic Regression:** interpretable linear model with class weighting.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, roc_auc_score, precision_recall_curve, auc
```

```python
lr = LogisticRegression(max_iter=1000, class_weight='balanced')
lr.fit(X_res, y_res)
```

```python
lr_pred = lr.predict(X_test)
lr_pred
```

```python
lr_prob = lr.predict_proba(X_test)[:,1]
```

```python
lr_prob
```

#### **Random Forest:** tree-based ensemble capturing nonlinear relationships.

```python
rf = RandomForestClassifier(n_estimators=100, max_depth=8, random_state=42)
rf.fit(X_res, y_res)
rf_pred = rf.predict(X_test) 
rf_prob = rf.predict_proba(X_test)[:,1]
```

```python
for model_name, pred, prob in [('Logistic Regression', lr_pred, lr_prob)
                               ,('Random Forest Classifier', rf_pred, rf_prob)]:
    print(f"\n{model_name} Results:")
    print(classification_report(y_test, pred))
    print("ROC-AUC:", roc_auc_score(y_test, prob))
```

Both models achieved strong overall separability (ROC-AUC â‰ˆ 0.97 â€“ 0.98).  
However, **Random Forest** provided the best fraud-detection trade-off:

| Model | Precision (Fraud = 1) | Recall (Fraud = 1) | F1 | ROC-AUC |
|--------|-----------------------|--------------------|----|---------|
| Logistic Regression | 0.05 | 0.92 | 0.10 | 0.97 |
| Random Forest | 0.28 | 0.89 | 0.43 | 0.98 |

- **Logistic Regression:** very high recall but poor precision â†’ too many false positives.  
- **Random Forest:** maintains high recall with better precision, meaning fewer false alarms.  

In fraud detection, recall is prioritized, so Random Forest currently offers the most balanced and practical model for production use.
